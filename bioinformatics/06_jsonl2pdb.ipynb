```python
import biotite
import biotite.structure as struc
import biotite.structure.io as strucio

# write a function convert 1 to 3 letter code

restype_1to3 = {
    'A': 'ALA',
    'R': 'ARG',
    'N': 'ASN',
    'D': 'ASP',
    'C': 'CYS',
    'Q': 'GLN',
    'E': 'GLU',
    'G': 'GLY',
    'H': 'HIS',
    'I': 'ILE',
    'L': 'LEU',
    'K': 'LYS',
    'M': 'MET',
    'F': 'PHE',
    'P': 'PRO',
    'S': 'SER',
    'T': 'THR',
    'W': 'TRP',
    'Y': 'TYR',
    'V': 'VAL',
}

def aggregra(data):
    coords_biotite = []
    sequence = data['seq']
    coords = data['coords']
    length = len(coords['N'])
    for i in range(length):
        atom_N = struc.Atom(
            atom_name="N",
            coord=coords['N'][i],
            res_id=i+1,
            chain_id="A",
            res_name=restype_1to3[sequence[i]],
            element="N",
        )
        atom_CA = struc.Atom(
            atom_name="CA",
            coord=coords['CA'][i],
            res_id=i+1,
            chain_id="A",
            res_name=restype_1to3[sequence[i]],
            element="C",
        )
        atom_C = struc.Atom(
            atom_name="C",
            coord=coords['C'][i],
            res_id=i+1,
            chain_id="A",
            res_name=restype_1to3[sequence[i]],
            element="C",
        )
        atom_O = struc.Atom(
            atom_name="O",
            coord=coords['O'][i],
            res_id=i+1,
            chain_id="A",
            res_name=restype_1to3[sequence[i]],
            element="O",
        )
        atom_CB = struc.Atom(
            atom_name="CB",
            coord=coords['CB'][i],
            res_id=i+1,
            chain_id="A",
            res_name=restype_1to3[sequence[i]],
            element="C",
        )
        coords_biotite.append(atom_N)
        coords_biotite.append(atom_CA)
        coords_biotite.append(atom_C)
        coords_biotite.append(atom_O)
        coords_biotite.append(atom_CB)
    return struc.array(coords_biotite)
```
