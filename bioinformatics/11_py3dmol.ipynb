```python
import py3Dmol

def py3dmol_render(view,row=0,column=0,method='plddt',system=None):
    if method == 'plddt':
        view.setStyle({'cartoon': {'colorscheme': {'prop':'b','gradient': 'roygb','min':50,'max':90}}},viewer=(row,column))
    elif method == 'spectrum':
        view.setStyle({'cartoon': {'color':'spectrum'}},viewer=(row,column))
    elif method == 'show_side_chains':
        # from colabfold
        BB = ['C','O','N']
        view.addStyle({'and':[{'resn':["GLY","PRO"],'invert':True},{'atom':BB,'invert':True}]},
                        {'stick':{'colorscheme':f"WhiteCarbon",'radius':0.3}},viewer=(row,column))
        view.addStyle({'and':[{'resn':"GLY"},{'atom':'CA'}]},
                            {'sphere':{'colorscheme':f"WhiteCarbon",'radius':0.3}},viewer=(row,column))
        view.addStyle({'and':[{'resn':"PRO"},{'atom':['C','O'],'invert':True}]},
                            {'stick':{'colorscheme':f"WhiteCarbon",'radius':0.3}},viewer=(row,column))
    elif method == 'show_main_chains':
        BB = ['C','O','N','CA']
        view.addStyle({'atom':BB},{'stick':{'colorscheme':f"WhiteCarbon",'radius':0.3}},viewer=(row,column))

n_row = 1
n_col = 1
size = 400
width = n_col * size
height = n_row * size
view = py3Dmol.view(js='https://3dmol.org/build/3Dmol.js',viewergrid=(n_row,n_col),width=width,height=height,linked=False)


pdb_path = f"/home/sirius/PhD/software/AF_Cluster/zjj_output/Abeta40/EX_000.pdb"
with open(pdb_path,'r') as f:
    system = f.read()
view.addModel(system,'pdb',viewer=(0,0))
py3dmol_render(view,row=0,column=0,method="spectrum")
view.addSurface(py3Dmol.VDW,{'opacity':0.7,'color':'blue'}, {'chain':'A'},viewer=(0,0))

# view.addLabel("F4",{'fontOpacity':0.5},{'resi':4})

# view.spin("y")
view.zoomTo()
view.show()

# view.zoomTo()
```
